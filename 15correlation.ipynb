# 상관분석


두 변수가 어떤 관계에 있는지 분석하는 과정
보통 2개의 등간/비율 척도변수를 분석할 대 우선적으로 산점도를 그려 변수간의 관계를 시각화 한 후 상관계수를 계산.


```
# A회사의 광고비 지출이 매출에 어느정도 영향을 주는지 파악하기 위해 광고비와 매출액간의 관계를 조사해 봄
# 광고비와 매출액간의 연관성 파악을 위해 산포도로 시각화해 보세요
ad <- c(13,8,10,15,12,15,14,15,17,19,20,21,22,21,25)
sales <- c(94,70,90,100,95,100,85,95,105,105,110,105,104,105,121)
plot(ad, sales)
```

```
# 상관관계
# 공분산 : 두 변수가 각 평균으로부터 떨어져 있는 정도를 구한 후 표본의 갯수로 나누어 계산한 것
# 양의 상관관계
# 음의 상관관계
```

```
# 앞의 예제를 참고해서 광고비와 매출액간의 수익성 변화를 알아보시오.
plot(ad, sales)
abline(v=mean(ad), col='red')
abline(h=mean(sales), col='blue')
```

```
# 판매원 15명의 판촉횟수와 판매량 자료가 다음과 같을 때 전화횟수에 따라 판매량 변화가 있는지 알아보세요.
p <- c(96,40,104,128,164,76,72,80,36,84,180,132,120,44,84)
s <- c(41, 41,51,60,61,29,39,50,28,43,70,56,45,31,30)
plot(p, s)
abline(v=mean(p), col='red')
abline(h=mean(s), col='blue')
```

```
# 공분산, 상관계수는 cov, cor 함수를 사용
# cor함수에는 method라는 속성이 지원
# pearson (기본) : 등간/비율 척도, 정규성
# spearman : 순서척도, 비정규성
# kendall : 순서척도, 비정규성

# 광고비와 매출액 간의 관계를 상관계수로 표현
cor(ad, sales)
# 전화판촉비와 판매량 간의 관계를 상관계수로 표현
cor(p, s)
```

```
# applewood 자동차판매 데이터를 기초로 구매자 연령과 판매이익간의 상관계수를 계산하고 상관분석을 하시오.
aw <- read.csv('https://raw.githubusercontent.com/mermerm/r_statistics/master/datasets/applewood.txt', header=T, sep=' ', na.strings = )
age <- aw$Age
profit <- aw$Profit
library(stringr)
profit <- str_remove_all(profit, '\\$') # 특문이라 \\를 삽입
profit <- str_remove_all(profit, ',')
profit <- as.integer(profit)  # 숫자형으로 바꿈
plot(age, profit)
abline(v=mean(age), col='red')
abline(h=mean(profit), col='blue')
cor(age, profit)
```

```
# 상관계수 범위에 따른 상관정도 해석
# -1.0 ~ -0.7 : 매우 강한 음의 상관관계
# -0.7 ~ -0.3 : 강한 음의 상관관계
# -0.3 ~ -0.1 : 약한 음의 상관관계
# -0.1 ~ 0.1 : 상관관계 없음
# 0.1 ~ 0.3 : 약한 양의 상관관계
# 0.3 ~ 0.7 : 강한 양의 상관관계
# 0.7 ~ 1.0 : 매우 강한 양의 상관관계
```

```
#  ggplot2가 제공하는 데이터셋 중 economics 데이터를 이용해서 실업자수unemploy, 개인소비지출pce, 개인저축율psavert의 상관관계를 분석하시오.
library(ggplot2)
str(economics)
unemploy <- economics$unemploy
pce <- economics$pce
save <- economics$psavert

plot(unemploy, pce)
abline(v=mean(unemploy), col='red')
abline(h=mean(pce), col='blue')
cor(unemploy, pce)

plot(unemploy, save)
abline(v=mean(unemploy), col='green')
abline(h=mean(save), col='yellow')
cor(unemploy, save)
```

```
# MASS 패키지의 Cars93 데이터셋을 이용해서 고속도로highway/시내city연비와 자체중량weight의 상관관계를 알아보세요
library(MASS)
high <- Cars93$MPG.highway
city <- Cars93$MPG.city
wt <- Cars93$Weight

plot(wt, high)
abline(v=mean(wt), col='red')
abline(h=mean(high), col='blue')
cor(wt, high)

plot(wt, city)
abline(v=mean(wt), col='yellow')
abline(h=mean(city), col='gray')
cor(wt, city)
```

```
 # 상관행렬
 # 여러 변수수들의 상관관례를 한번에 알아보고자 하는 경우 corrplot 함수를 이용하면 됨

#iris데이터셋에 대해 개별 변수 상관분석 시각화
plot(iris$Sepal.Length, iris$Sepal.Width)
plot(iris$Petal.Length, iris$Petal.Width)

# plot 함수를 이용해서 상관행렬 시각화
plot(iris[, 1:4])
```

```
# corrplot
# 상관관계정도를 한눈에 보기 쉽게 그래프로 그려줌
# 빨간색 : 음의 상관관계
# 파랑색 : 양의 상관관계
install.packages('corrplot')
library(corrplot)
cor <- cor(iris[, 1:4])
corrplot(cor, method='square')
corrplot(cor, method='pie')
corrplot(cor, method='shade')
corrplot(cor, method='ellipse')
corrplot(cor, method='number')
```

```
# MASS 패키지의 Cars93 데이터셋을 이용해서 고속도로highway/시내city 연비와 관련된 변수들의 상관관계를 corrplot 함수로 확인해보세요.

# 1) 고속도로 연비 vs 가격, 엔진크기, 마력, RPM, 중량
hwmpg <- Cars93[, c(8, 5, 12:14, 25)]
plot(hwmpg)
pairs(hwmpg, panel=panel.smooth)  # 그래프에 추세선 출력

cor <- cor(hwmpg)
cor # 상관계수 행렬

corrplot(cor, method='ellipse')
corrplot(cor, method='number')
```

# 상관계수 유의성 검정

```
# 독립변수와 종속변수가 서로 관련없는데 상관계수가 양/음으로 나온것은 아닌지?
# 유의수준 0.05를 기준으로 가설검정 실시
# R에서는 cor.test함수 사용
# 상관계수의 검정통계량은 t분포를 이용

# 광고비와 매출액에 대한 상관계수의 유의성 검정
options(scipen=100)
cor.test(ad, sales)
# 유의확률이 유의수준보다 작으므로 대립가설 채택
```

```
# 어느 대학교 학생회에서 학생들이 마시는 맥주와 혈중알콜농도 사이 관계를 알고자 한다. 18명의 학생을 표본추출하여 맥주캔 500ml을 마시게 하고 30분후 경찰서에서 혈중알콜농도를 측정하였음. 시음량과 알콜농도간의 상관관계를 알아보세요.

beer <- c(6, 7, 7, 4, 5, 3, 3, 6, 6, 3, 3, 7, 1, 4, 2, 7, 2, 1)
blood <- c(0.1, 0.09, 0.09, 0.1, 0.1, 0.07, 0.1, 0.12, 0.09, 0.07, 0.05, 0.08, 0.04, 0.07, 0.06, 0.12, 0.05, 0.02)

plot(beer, blood)
abline(v=mean(beer), col='red')
abline(h=mean(blood), col='blue')

df <- data.frame(beer, blood)
pairs(df, panel=panel.smooth)
cor <- cor(df)
cor
corrplot(cor, method='ellipse')
corrplot(cor, method='number')
chart.Correlation(df, histgram=T, pch=19)
cor.test(beer, blood)
```
