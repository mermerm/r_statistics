```
# 서울 교통사고 데이터 시각화
car <- read.csv('https://raw.githubusercontent.com/mermerm/r_statistics/master/datasets/seoul_car_2015.txt', sep=' ')
head(car)

junnggu <- car %>% filter(자치구명=='중구') %>% 
           select(월, 발생건수)
plot(junnggu$월, junnggu$발생건수, type='b', pch=15, ylim=c(0, 150), main='2015년 중구지역의 월별 교통사고 발생건수', xlab='월', ylab='발생건수')
```
